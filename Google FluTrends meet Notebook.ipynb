```python
data = pandas.io.parsers.read_csv('https://www.google.org/flutrends/%s/data.txt' % COUNTRY_CODE,
                                  skiprows=11)
data.set_index('Date', inplace = True)

data_to_display = None
if COUNTRY_CODE=='us':
  data_to_display = data.filter(like="HHS", axis=1)  # We only show HHS region.
else:
  data_to_display = data
```

```python
data_to_display.tail(5)                            # We show data for last 5 weeks.
data_to_display.tail(52).plot(figsize=(16, 8))     # We plot the data for the last year.
```
